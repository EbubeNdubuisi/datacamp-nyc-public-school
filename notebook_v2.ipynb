# Which NYC schools have the best math results?
The best math results are at least 80% of the *maximum possible score of 800* for math

80% of 800 = 0.8 * 800 = 640

Schools with average math scores >= 640

```python
# subset the schools dataframe for columns school_name and average math, and apply condition to the average_math column
best_math_schools = schools[schools["average_math"] >= 640][["school_name", "average_math"]]
# sort the subsetted dataframe
best_math_schools = best_math_schools.sort_values("average_math", ascending=False)
# Print the result
print (best_math_schools)
```

# What are the top 10 performing schools based on the combined SAT scores?

```python
# create the new column total_SAT consisting of the total average SAT scores
schools["total_SAT"] = schools["average_math"] + schools["average_reading"] + schools["average_writing"]
schools.head()
```

```python
# sort the schools df based on average total SAT scores
schools_sorted = schools.sort_values("total_SAT", ascending=False)
print (schools_sorted)
```

```python
# Subset the schools dataframe for the columns
sub_schools = schools_sorted[["school_name", "total_SAT"]]
print (sub_schools)
```

```python
# extract the top 10 from the sorted schools dataframe
top_10_schools = sub_schools.head(10)
print (top_10_schools)
```

# Which single borough has the largest standard deviation in the combined SAT score?

```python
import numpy as np

# group the schools df based on borough
grouped_schools = schools.groupby("borough").agg(
    num_schools=('school_name', 'count'),
    average_SAT=('total_SAT', 'mean'),
    std_SAT=('total_SAT', 'std')
).round(2)
print (grouped_schools)
```

```python
largest_std_dev = grouped_schools[grouped_schools["std_SAT"] == grouped_schools["std_SAT"].max()]
print (largest_std_dev)
```
